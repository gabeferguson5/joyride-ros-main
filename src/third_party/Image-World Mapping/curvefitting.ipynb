```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import lsq_linear
from mpl_toolkits.mplot3d import Axes3D
```

```python
U, V, X, Y, Z= np.loadtxt('positions.csv',delimiter=',',skiprows=1,unpack=True)
A = np.transpose([np.ones_like(U), U, U**2, U**3, V, V**2, V**3, U*V, U*V**2, U**2*V])
```

Φ(U,V) Θ  = [X,Y,Z]

```python
res = lsq_linear(A,X)
print(res)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(U,V,X,marker='o',label="data")
ax.scatter(U,V,np.dot(A,res.x),marker='x',label='fitted')
plt.title("(U,V) -> X")
ax.set_xlabel('u - [pix]')
ax.set_ylabel('v - [pix]')
ax.set_zlabel('x - [in]')
plt.legend()
plt.show()
```

```python
res = lsq_linear(A,Y)
print(res)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(U,V,Y,marker='o',label="data")
ax.scatter(U,V,np.dot(A,res.x),marker='x',label='fitted')
plt.title("(U,V) -> Y")
ax.set_xlabel('u - [pix]')
ax.set_ylabel('v - [pix]')
ax.set_zlabel('y - [in]')
plt.legend()
plt.show()
```
